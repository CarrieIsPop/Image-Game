Trabajando con imagenes

```python
import cv2
import matplotlib.pyplot as plt
path = "/content/Paisaje.jpg"
img = cv2.imread(path)

imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(imgRGB)
plt.show()

##############################################################

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
colored = cv2.applyColorMap(gray, cv2.COLORMAP_OCEAN)

colored_rgb = cv2.cvtColor(colored, cv2.COLOR_BGR2RGB)
plt.imshow(colored_rgb)
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import os
import random

folder_path = "/content"

image_files = [f for f in os.listdir(folder_path) if f.lower().endswith('.jpg')]

if len(image_files) < 5:
    print(" Necesitas al menos 5 imgenes en la carpeta para jugar.")
else:
    score = 0
    rounds = 5

    for i in range(rounds):
        print(f"\n Ronda {i+1} de {rounds}")

        chosen_file = random.choice(image_files)
        chosen_path = os.path.join(folder_path, chosen_file)

        img = cv2.imread(chosen_path)
        imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        guess = input("Qu imagen crees que va a aparecer? Escribe el nombre sin '.jpg': ").strip().lower()

        plt.imshow(imgRGB)
        plt.axis('off')
        plt.show()

        correct_name = os.path.splitext(chosen_file)[0].lower()
        if guess == correct_name:
            print(" Correcto! +10 puntos")
            score += 10
        else:
            print(f" Incorrecto. Era: {correct_name}. +0 puntos")

    print("\n Juego terminado")
    print(f"Tu puntuacin final fue: {score} de {rounds * 10}")
```
